```python
test

for i in test:
    print(i)

# for ind in test.index:     
#     feature1Probability = 
#     print(test['Feature1'][ind], test['Feature3'][ind]) 

train['CallCategory'].unique()
PostQ = train[train['CallCategory'] == "PostQ"]
PreQ = train[train['CallCategory'] == "PostQ"]
ProbabilityofPostQ = len(PostQ)/len(train)
ProbabilityofPreQ = len(PreQ)/len(train)
Feature1PostQ = {}
Feature2PostQ = {}
Feature3PostQ = {}
Feature4PostQ = {}

#PreQ Features
Feature1PreQ = {}
Feature2PreQ = {}
Feature3PreQ = {}
Feature4PreQ = {}

for i in PostQ['Feature1'].unique():
    Feature1PostQ[i] = len(PostQ[PostQ['Feature1'] == i])/len(PostQ)

for i in PostQ['Feature2'].unique():
    Feature2PostQ[i] = len(PostQ[PostQ['Feature2'] == i])/len(PostQ)


for i in PostQ['Feature3'].unique():
    Feature3PostQ[i] = len(PostQ[PostQ['Feature3'] == i])/len(PostQ)

for i in PostQ['Feature4'].unique():
    Feature4PostQ[i] = len(PostQ[PostQ['Feature4'] == i])/len(PostQ)
    
    
Feature1PostQ, Feature2PostQ, Feature3PostQ, Feature4PostQ


for i in PreQ['Feature1'].unique():
    Feature1PreQ[i] = len(PreQ[PreQ['Feature1'] == i])/len(PreQ)

for i in PreQ['Feature2'].unique():
    Feature2PreQ[i] = len(PreQ[PreQ['Feature2'] == i])/len(PreQ)


for i in PreQ['Feature3'].unique():
    Feature3PreQ[i] = len(PreQ[PreQ['Feature3'] == i])/len(PreQ)

for i in PreQ['Feature4'].unique():
    Feature4PreQ[i] = len(PreQ[PreQ['Feature4'] == i])/len(PreQ)
    
    

onlyFeature1 = {}
onlyFeature2 = {}
onlyFeature3 = {}
onlyFeature4 = {}


for i in train['Feature1'].unique():
    onlyFeature1[i] = len(train[train['Feature1'] == i])/len(train)

for i in train['Feature2'].unique():
    onlyFeature2[i] = len(train[train['Feature2'] == i])/len(train)


for i in train['Feature3'].unique():
    onlyFeature3[i] = len(train[train['Feature3'] == i])/len(train)

for i in train['Feature4'].unique():
    onlyFeature4[i] = len(train[train['Feature4'] == i])/len(train)
    

print(Feature1PostQ, Feature2PostQ, Feature3PostQ, Feature4PostQ)
print('\n')
print(Feature1PreQ, Feature2PreQ, Feature3PreQ, Feature4PreQ)
print('\n')
print(onlyFeature1,onlyFeature2,onlyFeature3,onlyFeature4)


```
